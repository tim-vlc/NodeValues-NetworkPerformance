```python
%matplotlib inline
import itertools
import numpy as np
import matplotlib.pyplot as plt
import math
import os
import pickle
from scipy.optimize import curve_fit
```

```python
cwd = os.getcwd()
path = cwd + '/results_composed.txt'
data = np.loadtxt(path)
```

```python
# create a boolean mask based on your conditions
mask = (data[:, 2] >= 0) & (data[:, 3] < 20)

# use the boolean mask to select the values
avg_dist = data[mask, 4]
avg_out = data[mask, 3]
test_err = data[mask, 2]
```

```python
# calculate polynomial
z_dist = np.polyfit(avg_dist, test_err, 1)
f_dist = np.poly1d(z_dist)

# calculate new x's and y's
x_dist = np.linspace(min(avg_dist), max(avg_dist), 100)
y_dist = f_dist(x_dist)

plt.plot(x_dist, y_dist, color='blue')
plt.plot(avg_dist, test_err, '.', color='blue')
plt.show()
```

```python
z_out = np.polyfit(avg_out, test_err, 1)
f_out = np.poly1d(z_out)

x_out = np.linspace(min(avg_out), max(avg_out), 100)
y_out = f_out(x_out)

plt.plot(x_out, y_out, color='orange')
plt.plot(avg_out, test_err, '.', color='orange')
```
